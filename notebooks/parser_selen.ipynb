```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from pyvirtualdisplay import Display
from xvfbwrapper import Xvfb
from selenium.webdriver.common.by import By
import pandas as pd
import time

def correct_url(url): 
    if not url.startswith("http://") and not url.startswith("https://"):
        url = "http://" + url
    return url

def scrollDown(browser, numberOfScrollDowns):
    body = browser.find_element(By.TAG_NAME,'body')
    print(body)
    button = browser.find_element(By.XPATH,'/html/body/div[2]/div[3]/div[3]/div/div/div[1]/div/div[2]/div/div/div')
    
    while numberOfScrollDowns >=0:
        body.send_keys(Keys.PAGE_DOWN)
        browser.execute_script("arguments[0].click();", button);
        numberOfScrollDowns -= 1
        print(numberOfScrollDowns)
        
    return browser
    

def crawl_url(url, run_headless=True):
    
    if run_headless:
        display = Display(visible=0, size=(1024, 768))
        display.start()
    
    url = correct_url(url)
    print(url)
    browser = webdriver.Firefox()
    browser.get(url)
    browser = scrollDown(browser, 300)
    
    
    
    all_hover_elements = browser.find_elements(By.CSS_SELECTOR, ".layout__content")
    print(all_hover_elements)
    
    
    herfs = []
    title_1 = []
    text_1 = []
    date_1 = [] 
    
    
    for hover_element in all_hover_elements:
        
        a_element = hover_element.find_elements(By.CSS_SELECTOR, "html body div.layout div.layout__wrapper div.layout__content.layout__content_section div.layout__rows div.rows div.rows__column.rows__column_section-left div.column div.column__row div.listing.listing_all-new.listing_js.listing-nocover div.listing__content.listing__content_js ul.listing__rows.listing__rows_js li.listing__column.listing__column_all-new.listing__column_js div.listing__card div.card.card_all-new div.card__heading.card__heading_all-new a.link.link_color")
        
        herf = [a_element.get_attribute('href') for a_element in a_element]
        
        
        herfs.append(herf)
        
        print(len(herfs[0]))
        
        try:
            for i in range(900, len(herfs[0])):

                print(i)
                print(herf[i])

                browser.get(herf[i])



                title = browser.find_element(By.CSS_SELECTOR, '.article__heading')
                print(title.text)
                title_1.append(title.text)

                text = browser.find_element(By.CSS_SELECTOR, '.article__text')
                print(text.text)
                text_1.append(text.text)

                date = browser.find_element(By.CSS_SELECTOR, '.date')
                print(date.text)
                date_1.append(date.text)
                time.sleep(3)
        except:
            print('error')
    dicts = {
        'herf':[herfs][0][0],
        'tilte':[title_1][0],
        'text':[text_1][0],
        'date_1':[date_1][0]
    }
    
    min_length = 3000
    data = pd.DataFrame(dicts)
    
    return data.to_csv('bus_2122_rt.csv')
    
    
    

url = "https://russian.rt.com/tag/biznes"
crawl_url(url)
```
